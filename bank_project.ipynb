```python
# Convert 'Age' column to numeric type. If there are any non-numeric values,
# they will be converted to NaN.
DF["Age"] = pd.to_numeric(DF["Age"], errors='coerce')

# Now you can calculate the median and fill NaN values
DF=DF.fillna(value={"Surname":"Missing Value","Age":DF["Age"].median()})
```
